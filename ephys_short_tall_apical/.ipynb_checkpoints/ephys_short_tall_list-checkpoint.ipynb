```python
v_colors={"Short": "#ef3b2c", "Tall": "#4292c6"}
```

```python
f, axes = plt.subplots(figsize=(20,7), nrows=1, ncols=5, sharex=True)

x="Projection Class"

ax0 = sns.boxplot(x=x, y="vrest", data=merged, hue="Morpho type",
                  hue_order=v_colors, palette=v_colors, saturation=1, ax=axes[0])
ax1 = sns.boxplot(x=x, y="ri", data=merged, hue="Morpho type",
                  hue_order=v_colors, palette=v_colors, saturation=1, ax=axes[1])
ax2 = sns.boxplot(x=x, y="upstroke_downstroke_ratio_long_square", data=merged, hue="Morpho type",
                  hue_order=v_colors, palette=v_colors, saturation=1, ax=axes[2])
ax3 = sns.boxplot(x=x, y="f_i_curve_slope", data=merged, hue="Morpho type",
                 hue_order=v_colors, palette=v_colors, saturation=1, ax=axes[3])
ax4 = sns.boxplot(x=x, y="sag", data=merged, hue="Morpho type",
                  hue_order=v_colors, palette=v_colors, saturation=1, ax=axes[4])

ax0.set_xlabel("")
ax1.set_xlabel("")
ax2.set_xlabel("Projection Class")
ax3.set_xlabel("")
ax4.set_xlabel("")

ax0.get_legend().remove()
ax1.get_legend().remove()
#ax2.get_legend().remove()
ax3.get_legend().remove()
ax4.get_legend().remove()

ax0.set_ylabel("Vm(mV)")
ax1.set_ylabel("Input Resistance(MÎ©)")
ax2.set_ylabel("LS AP up/downstoke")
ax3.set_ylabel("F-I Curve")
ax4.set_ylabel("Sag Fraction")

sns.despine(bottom=False, right=True)
f.tight_layout()
plt.suptitle("IT vs. CT", y=1.05)
```
